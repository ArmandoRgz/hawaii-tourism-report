```python
import pandas as pd
```

```python
folder = "source_files"
filenames = ["hawaii-hotel_performance","hawaii-visitor_trends","hawaii-expenditure_patterns","hawaii-air_seats","hawaii-visitor_characteristics"]
```

```python
raw_frames = {}
clean_frames = {}
for fname in filenames:
    path = f"{folder}/{fname}.csv"
    raw_frames[fname] = pd.read_csv(path)
    cols = raw_frames[fname].columns.tolist()
    id_vars = [col for col in cols if not col[0].isdigit()]
    clean_frames[fname] = raw_frames[fname].melt(id_vars = id_vars, var_name = "period_date", value_name = "value")
    clean_frames[fname] = clean_frames[fname].iloc[:-4]
    clean_frames[fname]["period_date"] = pd.to_datetime(clean_frames[fname]["period_date"])
    clean_frames[fname]["value"] = clean_frames[fname]["value"]
```

```python
for frame in clean_frames:
    print(clean_frames[frame].info())
```
